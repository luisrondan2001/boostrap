```python
import matplotlib.pyplot as plt

x=[2016, 2017, 2018, 2019, 2020,2021]
y=[45, 50, 60, 30, 50, 80]

x2=[2016, 2017, 2018, 2019, 2020,2021]
y2=[30, 50, 60, 10, 70, 60]

plt.plot(x,y, color='green',  marker='o', linestyle='--' , label='ecuador')
plt.plot(x2,y2, color='blue',  marker='d', linestyle='--', label='Peru')
plt.title("Icremento de casos")
plt.show()
 

```

```python
#grafica de pie
import matplotlib.pyplot as plt

lenguajes=('python', 'R', 'C', 'JS')
slices=( 20, 30, 40, 10)
colores=('red', 'green', '#ADEEE2', 'yellow')
valores=(0, 0.1 ,0 ,0)

plt.pie(slices, colors=colores, labels=lenguajes, explode=valores )
plt.title('Lenguajes de programacion')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def l(x):
  return np.cos(x)
x=np.linspace(0, 20, 50)

plt.plot(x, l(x), color='green',  marker='o', linestyle='--')
plt.show()
```
